## Question 2
### Assume marks obtained in three courses by a number of students are available in ‘mark.txt’ file. Load ‘mark.txt’ file into a numpy array

```python
#question 2
arr2 = np.loadtxt("Mark.txt", skiprows=1, dtype=np.int64)
print(arr2)
```

**a. print max of mark1, mark2 and mark3**

```python
#question 2 part a
np.max(arr2, axis=0)
```

**b. Plot mark1 and mark2 separately using a scatterplot. Also do this for mark1 and mark3. Observe for linearity between the marks and report.**

```python
#question 2 part b
import matplotlib.pyplot as plt
plt.scatter(arr2[:,0],arr2[:,1])
plt.title("Mark1 vs Mark2")
plt.xlabel("Mark1")
plt.ylabel("Mark2")
plt.show()
```
